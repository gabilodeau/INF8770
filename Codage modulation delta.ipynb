Exemple de codage modulation delta

```python
import numpy as np
import matplotlib.pyplot as plt
```

Création d'un signal à coder

```python
x = np.arange(1.0,360.0, 0.5)/360*2*np.pi
signal = abs(np.sin(x) + np.sin(2*x) + np.cos(3*x))*100
plt.plot(x, signal)
plt.show()
```

Codage modulation delta avec +/- delta. Un seul bit par échantillon... Si vous comparez avec l'exemple PCM, vous constaterez qu'il faut environ 6 bits pour bien représenter le signal original.

```python
Valprec = signal[0]
signalcode =[]  #signal de booléens
signaldelta = [Valprec]  #le signal reconstruit à l'aide des booléens
delta = 10  #Peut être changé
for i in range(1,len(signal)):
    if signal[i] > Valprec: #Augmente la force
        signalcode += [True]
        Valprec += delta
    else:                  #Diminue la force
        signalcode += [False]
        Valprec -= delta
    signaldelta += [Valprec]
```

```python
plt.figure(figsize = (15,6)) #On agrandit la figure pour bien voir les 2 courbes.
plt.plot(x, signaldelta,'b')
plt.plot(x, signal,'r')
plt.show()
```

```python
print('Signal codé: ', signal[0],signalcode)
```
